### 2. 정규 표현식 실습

```python
import re
```

```python
# 1) . 기호
r = re.compile("a.c")
r.search("kkk") # 아무런 결과도 출력 x
```

```python
r.search("abc")
```

```python
# 2) ? 기호
r = re.compile("ab?c")
r.search("abbc")
```

```python
r.search("abc")
```

```python
r.search("ac")
```

```python
# 3) * 기호
r = re.compile("ab*c")
r.search("a")
```

```python
r.search("ac")
```

```python
r.search("abc")
```

```python
r.search("abbbbc")
```

### 3. 정규 표현식 모듈 함수 예제

```python
# (1) re.match()와 re.search()의 차이
r = re.compile("ab.")
r.match("kkabc")
```

```python
r.search("kkkabc")
```

```python
r.match("abckkk")
```

### 4. 정규 표현식 텍스트 전처리 예제

```python
from nltk.tokenize import RegexpTokenizer

text = "Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"

tokenizer1 = RegexpTokenizer("[\w]+")
tokenizer2 = RegexpTokenizer("[\s]+", gaps=True)

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text))
```
